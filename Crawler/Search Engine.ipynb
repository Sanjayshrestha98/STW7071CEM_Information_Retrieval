```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()

    # Remove special characters, punctuation, and numbers
    text = re.sub(r"[^a-zA-Z\s]", "", text)

    # Tokenize the text into individual words
    words = text.split()

    # Remove stop words
    stop_words = set(stopwords.words("english"))
    words = [word for word in words if word not in stop_words]

    # Stem and lemmatize words
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(stemmer.stem(word)) for word in words]

    # Join the processed words back into a single string
    processed_text = " ".join(words)

    return processed_text

# Example usage
document = "I love running and playing soccer! 123"
processed_document = preprocess_text(document)
print(processed_document)
```

```python
def removeStopWord(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Remove special characters, punctuation, and numbers
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    
    # Tokenize the text into individual words
    words = text.split()
    
    words = [word for word in words if word not in stop_words]
    
    # Stem and lemmatize words
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(stemmer.stem(word)) for word in words]
    
    
    # Join the processed words back into a single string
    processed_text = " ".join(words)
    
    return processed_text
```

```python
from collections import defaultdict
def construct_inverted_index(documents):
    inverted_index = defaultdict(list)
    
    for doc_id, doc in enumerate(documents):
        terms = doc.split()
        
        # Update the inverted index for each term in the document
        for term in terms:
            inverted_index[term].append(doc_id)
    
    return inverted_index
```

```python
# Construct the inverted index
inverted_index = construct_inverted_index(pub_list_after_stem)
```

```python
#tfidf calculation
import math

def calculate_tf(term, document):
    term_frequency = document.count(term)
    return term_frequency / len(document.split())

def calculate_idf(term, inverted_index, total_documents):
    if term in inverted_index:
        document_frequency = len(inverted_index[term])
        return math.log(total_documents / document_frequency)
    return 0

def calculate_tfidf(term, document, inverted_index, total_documents):
    tf = calculate_tf(term, document)
    idf = calculate_idf(term, inverted_index, total_documents)
    return tf * idf


def calculate_document_vector(document, inverted_index, total_documents):
    document_vector = []
    for term in inverted_index.keys():
        tfidf = calculate_tfidf(term, document, inverted_index, total_documents)
        document_vector.append(tfidf)
    return document_vector

def calculate_cosine_similarity(query_vector, document_vector):
    dot_product = sum(query_term * document_term for query_term, document_term in zip(query_vector, document_vector))
    query_norm = math.sqrt(sum(query_term ** 2 for query_term in query_vector))
    document_norm = math.sqrt(sum(document_term ** 2 for document_term in document_vector))
    if query_norm == 0 or document_norm == 0:
        return 0
    return dot_product / (query_norm * document_norm)
```

```python
def search(query, inverted_index, documents, threshold):
    # Preprocess the query
    processed_query = removeStopWord(query)
    query_terms = processed_query.split()

    # Calculate TF-IDF scores for query terms
    query_vector = [calculate_tfidf(term, processed_query, inverted_index, len(documents)) for term in query_terms]

#     # Initialize a dictionary to store document scores
#     doc_scores = {}

     # Calculate document vectors
    document_vectors = [calculate_document_vector(doc, inverted_index, len(documents)) for doc in documents]
    
    # Calculate cosine similarity scores
    scores = [calculate_cosine_similarity(query_vector, doc_vector) for doc_vector in document_vectors]

    # Retrieve the relevant documents based on the threshold
    relevant_docs = [doc for doc, score in zip(documents, scores) if score > threshold]

    return relevant_docs
```
